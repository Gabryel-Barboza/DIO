```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho para o arquivo subido
df = pd.read_excel("SuperStoreUS-2015.xlsx")
```

```python
# Renomeando todas as colunas para o Português
df = df.rename(columns={"Row ID":"ID Linha", "Order Priority":"Prioridade da Ordem", "Discount":"Desconto", "Unit Price":"Preço Unitário",
                        "Shipping Cost":"Custo de Exportação", "Customer ID":"ID do Cliente", "Customer Name":"Nome do Cliente",
                        "Ship Mode":"Meio de Exportação", "Customer Segment":"Tipo de Cliente", "Product Category":"Categoria de Produto",
                        "Product Sub-Category":"Subcategoria de Produto", "Product Container":"Contêiner do Produto", "Product Name":"Nome do Produto",
                        "Product Base Margin":"Margem Base do Produto", "Country":"País", "Region":"Região", "State or Province":"Estado ou Província",
                        "City":"Cidade", "Postal Code":"Código Postal", "Order Date":"Data do Pedido", "Ship Date":"Data de Exportação",
                        "Profit":"Faturamento", "Quantity ordered new":"Quantidade", "Sales":"Receita", "Order ID":"ID do Pedido"})
print(df.columns)
# Verificando tipos de dados
print(df.dtypes)
df.shape
```

```python
df.head(7)
```

```python
# Preenchendo campos nulos
df = df.fillna(0)
df.isnull().sum()
```

```python
# Começando análise exploratória

# Alterando o estilo dos gráficos
plt.style.use("seaborn")
# Total de pedidos por mês
df.groupby(df['Data do Pedido'].dt.month).value_counts("").plot.bar(title="Total Pedidos por Mês")
plt.xlabel("Meses")
plt.ylabel("Quantidade");
```

```python
print("Tempo máximo para Envio: ", (df['Data de Exportação'] - df['Data do Pedido']).dt.days.max(), "dias")
print()
# Receita total por estado
receitas_estado = df.groupby(["Estado ou Província"])["Receita"].sum().reset_index()
# Estados com maiores vendas nos EUA (a partir de 50000)
maiores_receitas = receitas_estado.loc[receitas_estado["Receita"] > 50000].sort_values("Receita", ascending=False)
# Gráfico com as labels alteradas e rotação de 90 graus
maiores_receitas.plot.bar(title="Receita por Estado").set_xticklabels(maiores_receitas["Estado ou Província"], rotation=90);
```

```python
# Gráfico de Pizza com a categoria mais vendida, eixos desligados para retirar textos por cima do gráfico
df.groupby(["Categoria de Produto"])["Quantidade"].sum().plot.pie(title="Departamentos mais vendidos").axis("off");
```

```python
# Gráfico de linhas com o faturamento por mês
df.groupby(df["Data do Pedido"].dt.month)["Faturamento"].sum().plot(title="Faturamento por Mês", marker="o")
plt.xlabel("Mês")
plt.ylabel("Faturamento");
```
